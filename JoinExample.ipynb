***Join Types: left, right, inner, cross, self, outer join***

There is no difference in left and leftouter  likewise others right and rightouter

```
student = [
    (1, "Prakash", 1, "Bharatpur", 3.8), \
    (2, "Arun", 2, "Biratnagar", 4.0), \
    (3, "Shiv", 3, "Jhapa", 3.9), \
    (4, "Ramesh", 1, "Butwal", 4.0)
]
columns_student = ["ID", "StudentName", "Depart", "City","Marks"]

depart = [
    (1, "MECHANICAL"), \
    (2, "COMPUTER"), \
    (3, "CIVIL"), \
    (4, "ELECTRICAL")
]
columns_depart = ["ID", "departmentName"]
df1 = spark.createDataFrame(data = student, schema = columns_student)
df2 = spark.createDataFrame(data = depart, schema = columns_depart)
```

```
df1.show()
df2.show()
```

```
df1.join(df2,df1.Depart == df2.ID, 'inner').select(df1.ID,df1.StudentName,df1.Marks,df2.departmentName,).show()
```
